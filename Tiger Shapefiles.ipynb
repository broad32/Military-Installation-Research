## Tiger Shapefiles

These files come from the United States Census Bureau Repository containing geography files for Indian Reservations, Roads, Legislative Districts, and other sites of government interest.  I will pull each of the files available under the Military Installations section of the data, year by year.

Site: https://www.census.gov/cgi-bin/geo/shapefiles/index.php

```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from simpledbf import Dbf5
import geopandas as gpd
```

### 2017

```python
tiger17=Dbf5('tl_2017_us_mil.dbf')
tiger17=tiger17.to_dataframe()
tiger17.head()
```

```python
geotiger17= gpd.GeoDataFrame.from_file("tl_2017_us_mil.shp")
geotiger17.head()
```

```python
geotiger17['geometry'].describe()
```
